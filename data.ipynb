**Setting new columns**

<span style="font-size:14px;">
    Add new column automatically align by index<br>
</span>

```python
s1 = pd.Series(np.random.randint(1, 101, size=len(data)), index=range(0,len(data)))
print(s1)

data["Random Number"] = s1

data[data["Random Number"] == 100]
```

```python
data.at[0, "smoker"] = "Yes"

data.iat[0, 1] = 10

data.loc[0] 
```

```python
data.loc[:, "Random Number"] = np.array([5] * len(data))
# difference between pandas series and numpy array is the start index

data.head(3)
```
